```python
import pandas

pandas.options.display.max_columns = None
pandas.options.display.max_rows = None

netflix_titles_data = pandas.read_csv("netflix_titles.csv")

netflix_titles_data.head(10)
```

```python
''' What release year is most prominent on Netflix?

- separate data based on release year to see which release years exist in dataset
- determine which release year appears the most in the dataset

'''

# find all release years that exist in dataset
release_years = netflix_titles_data['release_year'].dropna().unique()
release_years
```

```python
# finds which release year appears most frequently in dataset

netflix_titles_data['release_year'].value_counts()
```

```python
''' What type of media has Netflix focused on releasing/adding more: movies or shows?

- separate data based on type of media
- find how many times each type appears in dataset
- compare the number of times that each type appears

'''

# finds the different types of media that are present on Netflix

netflix_titles_data['type'].unique()
```

```python
# finds how many times each type of media appears in dataset

movie_count = len(netflix_titles_data[netflix_titles_data['type'] == 'Movie'])
tv_show_count = len(netflix_titles_data[netflix_titles_data['type'] == 'TV Show'])

print('Movies: ' + str(movie_count))
print('TV Shows: ' + str(tv_show_count))
```

```python
''' What are the most prominent ratings?

- separate data into ratings
- find how many times each rating appears
- compare amount of each rating

'''

# finds the different ratings that are present on Netflix

netflix_titles_data['rating'].unique()
```

```python
# sorts the rating data into descending order of the amount of times each rating appears

rating_dict={}

for rating in list(netflix_titles_data['rating'].unique()):
    rating_dict[rating] = len(netflix_titles_data[netflix_titles_data['rating'] == rating])

sorted_ratings = sorted(rating_dict, key= rating_dict.get, reverse=True)

print(sorted_ratings[0] + ' Rating Count: ' + str(rating_dict[sorted_ratings[0]]))
print(sorted_ratings[1] + ' Rating Count: ' + str(rating_dict[sorted_ratings[1]]))
print(sorted_ratings[2] + ' Rating Count: ' + str(rating_dict[sorted_ratings[2]]))
```

```python
''' What genres are most prominent on Netflix?

- separate data into genres
- find how many times each genre appears
- sort the genre dat into descending order of the amount of times each genre appears
'''

# finds the different genres that are present on Netflix

netflix_titles_data['listed_in'].unique()
```

```python
# sorts the genra data into descending order of the amount of times each genre appears

genre_dict={}

for genre in list(netflix_titles_data['listed_in'].unique()):
    genre_dict[genre] = len(netflix_titles_data[netflix_titles_data['listed_in'] == genre])

sorted_genres = sorted(genre_dict, key= genre_dict.get, reverse=True)

print(sorted_genres[0] + ' Genre Count: ' + str(genre_dict[sorted_genres[0]]))
print(sorted_genres[1] + ' Genre Count: ' + str(genre_dict[sorted_genres[1]]))
print(sorted_genres[2] + ' Genre Count: ' + str(genre_dict[sorted_genres[2]]))
```
