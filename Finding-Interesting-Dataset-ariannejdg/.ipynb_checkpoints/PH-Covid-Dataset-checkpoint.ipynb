```python
import pandas
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn

pandas.options.display.max_columns = None
pandas.options.display.max_rows = None

ph_covid_expenses_data = pandas.read_csv("covid_expenses1.csv")
ph_covid_expenses_data
```

# Q1: What is the range between the average expenses made before and during the pandemic?

```python
# get average expense pre-covid
avg_pre_covid_exp = int(ph_covid_expenses_data.iloc[:,[3]].sum()/len(ph_covid_expenses_data.iloc[:,[3]]))

# get average expense duringcovid
avg_covid_exp = int(ph_covid_expenses_data.iloc[:,[12]].sum()/len(ph_covid_expenses_data.iloc[:,[12]]))

#subtract avgs to find range
range_exp = (avg_pre_covid_exp - avg_covid_exp)
range_exp
```

# Q2: Who spent the most during the pandemic: men or women?

```python
# find avg of women's expenses
fem_data = ph_covid_expenses_data[ph_covid_expenses_data['Gender']=='Female']
avg_fem_exp = int(fem_data.iloc[:,[12]].sum()/len(ph_covid_expenses_data[ph_covid_expenses_data['Gender'] == 'Female']))
print("Women's expenses: ", avg_fem_exp)

# find avg of men's expenses
masc_data = ph_covid_expenses_data[ph_covid_expenses_data['Gender']=='Male']
avg_masc_exp = int(masc_data.iloc[:,[12]].sum()/len(ph_covid_expenses_data[ph_covid_expenses_data['Gender'] == 'Male']))
print("Men's expenses: ", avg_masc_exp)
```

# Q6: What is the relationship between age, transportation expenses, and expenses before the pandemic?

# Q10: Make a prediction on the expenses made during the pandemic based on one's age.

```python
from sklearn.linear_model import LinearRegression
x = ph_covid_expenses_data['Age'].values.reshape(-1,1)
y = ph_covid_expenses_data['Dur.Monthly Expenses ']

lr_model = LinearRegression()
lr_model.fit(x, y)
y_pred = lr_model.predict(x)

plt.scatter(x, y)
plt.xlabel('Age')
plt.ylabel('Dur.Monthly Expenses ')

plt.plot(x, y_pred)
```
