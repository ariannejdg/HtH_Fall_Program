```python
import pandas
import seaborn
import matplotlib.pyplot as plt

''' 
The best way to complete this activity is to copy the code in this file into a jupyter notebook file you create. This will also you to see your visualizations easier. Once you have finished creating your visuals copy your code back into this file and submit it to Replit. Don't forget to also upload your new code to Github! 
'''

# call in titanic csv
titanic_data = pandas.read_csv("titanic.csv")
titanic_data
```

```python
# Step 1: Create a heatmap showing the correlation between the following features: survival, pclass, sex, age, and fare 
titanic_features = titanic_data[['Survived', 'Pclass','Sex','Age','Fare']]
seaborn.heatmap(titanic_features.corr(), annot = True)
plt.show()
```

```python
# Step 2: Create a bar graph that illustrates how many passengers fell into each of these age ranges: 0 - 16, 17 - 25, 26 -40, 41 - 59, 60 or older

age_group_chart = {
    
    "groups": [
        "16 or younger",
        "17 to 25",
        "26 to 40",
        "41 to 60",
        "60+"
    ],
    
    "count": [
        len(titanic_data[ titanic_data["Age"] <= 16]),
        len(titanic_data[ (titanic_data["Age"] > 16) & (titanic_data["Age"] <= 25)]),
        len(titanic_data[ (titanic_data["Age"] > 25) & (titanic_data["Age"] <= 40)]),
        len(titanic_data[ (titanic_data["Age"] > 40 ) & (titanic_data["Age"] <= 59)]),
        len(titanic_data[ titanic_data["Age"] >= 60])
    ],
}

vis_df = pandas.DataFrame(age_group_chart)
seaborn.barplot(data = vis_df, x = 'groups', y = 'count');
plt.show()
```

```python
# Step 3: Create a line graph showing the average survival percentage
# of each of the previous age groups

#length of each list of filtered ages
children = len(titanic_data[(titanic_data["Age"] <= 16)])
ya = len(titanic_data[(titanic_data["Age"] >= 17) & (titanic_data["Age"] <= 25)])
millenial = len(titanic_data[(titanic_data["Age"] >= 26) & (titanic_data["Age"] <= 40)])
adults = len(titanic_data[(titanic_data["Age"] >= 41) & (titanic_data["Age"] <= 59)])
seniors = len(titanic_data[titanic_data["Age"] >= 60])

#Survived within age filter
children_survived = len(titanic_data[(titanic_data["Age"] <= 16) & (titanic_data["Survived"] == 1)])
ya_survived = len(titanic_data[(titanic_data["Age"] >= 17) & (titanic_data["Age"] <= 25)  & (titanic_data["Survived"] == 1)])
millenial_survived = len(titanic_data[(titanic_data["Age"] >= 26) & (titanic_data["Age"] <= 40)  & (titanic_data["Survived"] == 1)])
adults_survived = len(titanic_data[(titanic_data["Age"] >= 41) & (titanic_data["Age"] <= 59)  & (titanic_data["Survived"] == 1)])
seniors_survived = len(titanic_data[(titanic_data["Age"] >= 60)  & (titanic_data["Survived"] == 1)])

#Percentage of survival
children_percent =  children_survived/children*100
ya_percent = ya_survived/ya*100
millenial_percent = millenial_survived/millenial*100
adults_percent = adults_survived/adults*100
senior_percent = seniors_survived/seniors*100
ages = ['16 or younger','17-25','26-40', '41-59', '60+']
percents = [children_percent, ya_percent, millenial_percent, adults_percent, senior_percent]
age_groups = { #dict of lists
    'groups': ages, 
    'Survival Percent': percents
}

data_plot = pandas.DataFrame(age_groups)

seaborn.lineplot(data = data_plot, x = "groups", y = "Survival Percent", )
plt.show()

                    
```

```python
# Step 4: Create a pie chart that shows the percentage of survivors that were male and the percentage of survivors that were female
survivors = len(titanic_data[titanic_data['Survived'] == 1])
male_survival = len(titanic_data[(titanic_data['Survived'] == 1) & (titanic_data['Sex'] == 'male')])
male_survival_per = male_survival/survivors*100

female_survival = len(titanic_data[(titanic_data['Survived'] == 1) & (titanic_data['Sex'] == 'female')])
female_survival_per = female_survival/survivors*100

data = [male_survival_per, female_survival_per]
labels = ['Male survivors', 'Female survivors']

plt.pie(data, labels = labels, autopct='%.0f%%')
plt.show()
```

```python
# Step 5: Create a histogram that shows the distribution
# of passengers between the three embarking locations: 
# C (Cherbourg)[represented by 0], Q (Queenstown)[represented by 1], S (Southampton)[represented by 2]

embark_num = []

for _ , row in titanic_data.iterrows():
    
    if row["Embarked"] == "C":
        embark_num.append(0)
        
    if row["Embarked"] == "Q":
        embark_num.append(1)
        
    if row["Embarked"] == "S":
        embark_num.append(2)
        
    if pandas.isna(row["Embarked"]):
        embark_num.append(3)
        
titanic_data["embarked"] = embark_num   
titanic_data['embarked'].hist(bins=3)
```
