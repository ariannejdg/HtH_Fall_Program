```python
import pandas

pandas.options.display.max_columns = None
pandas.options.display.max_rows = None

netflix_titles_data = pandas.read_csv("netflix_titles.csv")

netflix_titles_data.head(10)
```

```python
'''What release year is most prominent on Netflix?

- separate data based on release year to see which release years exist in dataset
- determine which release year appears the most

'''

# find all release years that exist in dataset
release_years = netflix_titles_data['release_year'].dropna().unique()
release_years
```

```python
# finds which release year appears most frequently in dataset

netflix_titles_data['release_year'].mode()
```

```python
# shows an idea of how much of the dataset is composed of the most frequent release year

release_year_2018 = netflix_titles_data['release_year'] == 2018
release_year_2018
```

```python
# determines how many times the most frequent release year appears in dataset

released_2018 = netflix_titles_data[release_year_2018]
len(released_2018)
```

```python
'''
The most frequent release year in the dataset is 2018.
'''
```
